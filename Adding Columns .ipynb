```python
import pandas as pd
```

```python
import numpy as np
```

```python
df_exams= pd.read_csv('StudentsPerformance.csv')
```

```python
#Adding a Column with same value

df_exams['Language score']=70
df_exams
```

```python
#Adding a Column with values in a series 

language_score = np.arange(0,1000)

```

```python
#Adding a Column with random integer values

language_score= np.random.randint(1,100,size=1000)
```

```python
#Adding a Column with random float values b/w 1 to 100

language_score= np.random.uniform(1,100,size=1000)
Rlanguage_score=np.round(language_score,2) 
```

```python
df_exams['Language score']= Rlanguage_score
df_exams
```

# Adding Multiple Columns 

```python
#Using assign() (for overwriting data)
#Returns a new object (A copy) with all the original columns in addition to the new ones

#Create Random numbers for new column

Score1= np.random.randint(1,100,size=1000)
Score2= np.random.randint(1,100,size=1000)
```

```python
#Create series to specify positional argument (only 1)
Series1= pd.Series(Score1, index= np.arange(0,1000))
Series2= pd.Series(Score2, index= np.arange(0,1000))
```

```python
#Using assign() to add multiple columns
df_exams = df_exams.assign(Score1=Series1,Score2=Series2)
```

```python
#Show dataframe
df_exams
```

```python
#Insert() (used for inserting column at a particular index)

df_exams.insert(3,'fest',Series2)
df_exams
```
